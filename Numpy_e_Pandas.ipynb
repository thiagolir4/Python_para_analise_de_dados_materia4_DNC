#GroupBy

```python
import numpy as np
import pandas as pd
```

```python
dados = {
    "Nome": ["Ana", "Bruno", "Carla", "Daniel", "Eduardo", "Fernanda"],
    "Departamento": ["TI", "TI", "RH","RH", "VENDAS", "VENDAS"],
    "Salario": [5000, 6000, 4000, 4200, 3000, 3500],
    "Estadao": ["SP", "RJ", "SP", "RJ", "RJ", "SP"]
}
```

```python
df = pd.DataFrame(dados)
```

```python
df
```

```python
#faço o agrupamento por departamentos e depois seleciono a coluna na qual quero quero fazer as operações

df.groupby("Departamento")["Salario"].mean()
```

```python
df.groupby("Departamento")["Salario"].max()
```

```python
df.groupby("Departamento")["Salario"].count()
```

```python
#faz a contagem de valores unicos
df.groupby("Departamento")["Salario"].nunique()
```

```python
#analise agrupada das informações
df.groupby("Departamento").agg({"Salario": ["mean", "min", "max", "count", "nunique"]})
```

```python
import pandas as pd
import numpy as np

# Criando os dataframes

usuarios_df = pd.DataFrame({
"ID_USUARIO": [1, 2, 3],
'Nome': ['Maria', 'Bruno', 'João'],
'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte']
})

compras_df = pd.DataFrame({
    "ID_USUARIO": [1,2,1,1,3,3,2,3],
    "Produto": ["Notebook", "Camera", "Camiseta","Camiseta", "Caneca", "Caneca", "Calça", "Calça"],
    "Valor": [2000,1000,70,60,50,50,120,100]
})

novos_usuarios_df = pd.DataFrame({
'ID_USUARIO': [4, 5],
'Nome': ['Pedro', 'Marcos']
})
```

```python
usuarios_df
```

```python
compras_df
```

```python
novos_usuarios_df
```

```python
#unindo as df usuarios e novos_usuarios
usuarios_geral = pd.concat([usuarios_df, novos_usuarios_df])
```

```python
usuarios_geral
```

```python
#atualização feita somente nessa exibição, não fica fixo no df final
usuarios_geral.fillna({"Cidade": "Cidade Não Informada"})
```

```python
#para alocar isso no usuario final, devemos utilizar o argumento inplace = True no final
usuarios_geral.fillna({"Cidade": "Cidade Não Informada"}, inplace=True)
```

```python
usuarios_geral
```

```python
#usando o merge
#metodos de junção
#Left => manter todos os df da esquerda e associar os valores da df da direirta
#Right => vai associar todos os dados da direita com o da esquerda
#inner => vai associar e apresentar somente os dados que tiverem relação, como no caso dos df utilizados
#ele excluiu os dois novos usuarios que n tinha compras realizada

compras_df.merge(usuarios_geral, how="left", on="ID_USUARIO" )
```

```python
compras_df.merge(usuarios_geral, how="right", on="ID_USUARIO" )
```

```python
compras_df.merge(usuarios_geral, how="inner", on="ID_USUARIO" )
```
